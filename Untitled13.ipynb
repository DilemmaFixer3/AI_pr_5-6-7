```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, GRU, SimpleRNN, Dense, Dropout

# --- 1. Генерація Штучних Даних (Імітація traffic_data.csv) ---
np.random.seed(42)
time_steps = 24 * 365  # Один рік погодинних даних
time = np.arange(time_steps)
# Імітація добової сезонності (піки вранці та ввечері)
daily_pattern = (np.sin(time / 24 * 2 * np.pi * 2 + 10) + 1.5) * 100
# Імітація тижневої сезонності (менше трафіку у вихідні)
weekly_pattern = np.sin(time / (24 * 7) * 2 * np.pi) * 50 + 100
# Загальний тренд
trend = time / 100
# Трафік з шумом
traffic_volume = daily_pattern + weekly_pattern + trend + np.random.randn(time_steps) * 20

data = pd.DataFrame({'time': pd.to_datetime(pd.date_range(start='2023-01-01', periods=time_steps, freq='H')),
                     'traffic_volume': traffic_volume.astype(int)})

# Використовуємо лише стовпець трафіку для прогнозування
traffic_data = data['traffic_volume'].values.reshape(-1, 1)

print(f"Загальна кількість точок даних: {len(traffic_data)}")
```

```python
# --- 2.1. Нормалізація Даних ---
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(traffic_data)

# --- 2.2. Функція для Створення Послідовностей ---
def create_sequences(data, lookback=24, forecast_horizon=1):
    """Створює вибірки X (послідовності) та y (ціль)."""
    X, y = [], []
    for i in range(len(data) - lookback - forecast_horizon + 1):
        # X: послідовність довжиною lookback
        X.append(data[i:(i + lookback), 0])
        # y: ціль - значення через forecast_horizon кроків
        y.append(data[i + lookback + forecast_horizon - 1, 0])
    return np.array(X), np.array(y)

# Параметри послідовності
LOOKBACK = 24  # Використовуємо дані за останні 24 години
FORECAST_HORIZON = 1 # Прогнозуємо на 1 годину вперед

X, y = create_sequences(scaled_data, LOOKBACK, FORECAST_HORIZON)

# Перетворення форми X для RNN/LSTM/GRU: [зразки, lookback, ознаки]
X = X.reshape((X.shape[0], X.shape[1], 1))

# Розділення на навчальний та тестовий набори (80% / 20%)
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

print(f"Форма X_train: {X_train.shape}")
print(f"Форма X_test: {X_test.shape}")
```

```python
# --- 3.1. Функція для Побудови Моделі ---
def build_model(model_type, lookback):
    """Створює модель SimpleRNN, LSTM або GRU."""
    model = Sequential()

    if model_type == 'RNN':
        model.add(SimpleRNN(50, activation='relu', input_shape=(lookback, 1)))
    elif model_type == 'LSTM':
        # Використовуємо один LSTM шар
        model.add(LSTM(50, activation='relu', input_shape=(lookback, 1)))
    elif model_type == 'GRU':
        model.add(GRU(50, activation='relu', input_shape=(lookback, 1)))
    else:
        raise ValueError("Невідомий тип моделі")

    model.add(Dropout(0.2))
    model.add(Dense(1)) # Вихідний шар для прогнозу одного числа

    model.compile(optimizer='adam', loss='mse')
    return model

# --- 3.2. Функція для Оцінки ---
def evaluate_model(y_true_scaled, y_pred_scaled, scaler):
    """Перетворює назад та обчислює метрики."""
    # Зворотне перетворення
    y_true = scaler.inverse_transform(y_true_scaled.reshape(-1, 1))
    y_pred = scaler.inverse_transform(y_pred_scaled.reshape(-1, 1))

    # Обчислення метрик
    mse = mean_squared_error(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mse)

    return mse, mae, rmse, y_true, y_pred

# --- 3.3. Навчання та Оцінка Всіх Моделей ---
EPOCHS = 30
BATCH_SIZE = 32
model_types = ['RNN', 'LSTM', 'GRU']
results = {}

for model_type in model_types:
    print(f"\n===== Навчання моделі: {model_type} =====")

    model = build_model(model_type, LOOKBACK)

    # Навчання моделі
    history = model.fit(X_train, y_train,
                        epochs=EPOCHS,
                        batch_size=BATCH_SIZE,
                        verbose=0,
                        validation_split=0.1)

    # Прогноз на тестових даних
    y_pred_scaled = model.predict(X_test)

    # Оцінка
    mse, mae, rmse, y_true, y_pred = evaluate_model(y_test, y_pred_scaled, scaler)

    results[model_type] = {
        'MSE': mse,
        'MAE': mae,
        'RMSE': rmse,
        'y_true': y_true,
        'y_pred': y_pred
    }

    print(f"Результати для {model_type}:")
    print(f"  MSE: {mse:.2f}")
    print(f"  MAE: {mae:.2f}")
    print(f"  RMSE: {rmse:.2f}")
```

```python
# --- 4.1. Виведення Таблиці Результатів ---
results_df = pd.DataFrame({
    'Модель': list(results.keys()),
    'MSE': [res['MSE'] for res in results.values()],
    'MAE': [res['MAE'] for res in results.values()],
    'RMSE': [res['RMSE'] for res in results.values()]
})

print("\n===== Порівняння Метрик Прогнозування =====")
print(results_df.round(2).sort_values(by='MAE'))

# --- 4.2. Візуалізація Прогнозу ---
n_plot = 24 * 7 # Покажемо останній тиждень тестової вибірки
x_axis = np.arange(len(y_test) - n_plot, len(y_test))

plt.figure(figsize=(15, 8))
plt.plot(x_axis, results['LSTM']['y_true'][-n_plot:], label='Реальний Трафік', color='blue', linewidth=2)

for model_type in model_types:
    plt.plot(x_axis, results[model_type]['y_pred'][-n_plot:],
             label=f'Прогноз {model_type}', linestyle='--', alpha=0.7)

plt.title('Прогнозування Інтенсивності Трафіку (Порівняння Моделей)')
plt.xlabel('Час (години)')
plt.ylabel('Інтенсивність Трафіку')
plt.legend()
plt.grid(True)
plt.show()
```
